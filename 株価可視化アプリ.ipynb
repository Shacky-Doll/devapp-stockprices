```python
 #全ての銘柄に適用

days = 20
tickers = {
    "apple" : "AAPL",
    "meta" : "FB",
    "alphabet" : "GOOGL",
    "amazon" : "AMZN",
    "tesla" : "TSLA"
}

df = pd.DataFrame()

for company in tickers.keys():

    tkr = yf.Ticker(tickers[company])

    hist = tkr.history(period = f"{days}d")

    hist.index = hist.index.strftime("%d %B %Y")

    hist = hist[["Close"]]
    hist.rename(columns = {"Close" : tickers[company]}, inplace = True)

    hist = hist.T
    hist.index.name = "Ticker"
    df = pd.concat([df, hist], axis = 0)

df
```

```python
#関数化
def get_data(tickers, days):

    df = pd.DataFrame()

    for company in tickers.keys():

        tkr = yf.Ticker(tickers[company])

        hist = tkr.history(period = f"{days}d")

        hist.index = hist.index.strftime("%d %B %Y")

        hist = hist[["Close"]]
        hist.rename(columns = {"Close" : tickers[company]}, inplace = True)

        hist = hist.T
        hist.index.name = "Ticker"
        df = pd.concat([df, hist], axis = 0)

    return(df)
```

```python
get_data(tickers, days)
```

```python
import altair as alt
```

```python
df
```

```python
companies = ["AAPL", "FB"]
data = df.loc[companies]
data
```

```python
data = data.T.reset_index()
data.head()
```

```python
data = pd.melt(data, id_vars = ["Date"]).rename(columns = {"value" : "Stock Prices(USD)"})
data
```

```python
ymin = 180
ymax = 240

chart = (
    alt.Chart(data)
    .mark_line(opacity = 0.8, clip = True)
    .encode(x = "Date:T",
            y = alt.Y("Stock Prices(USD):Q", stack = None, scale = alt.Scale(domain = [ymin, ymax])),
            color = "Ticker:N"
    ) 
)
```
